## 1. Install and Import Required Libraries

First, we'll install the necessary packages and import the required modules.

```python
# Install required packages (uncomment if needed)
# !pip install crewai crewai-tools openai python-dotenv langchain-openai

import os
from dotenv import load_dotenv
from crewai import Agent, Task, Crew, Process
from langchain_openai import ChatOpenAI

print("âœ… All libraries imported successfully!")
```

## 2. Set Up OpenAI API Configuration

Configure the OpenAI API key and initialize the language model that will power our agents.

```python
# Load environment variables from .env file
load_dotenv()

# Get OpenAI API key from environment or set it directly
# IMPORTANT: Replace with your actual API key or use .env file
openai_api_key = os.getenv("OPENAI_API_KEY", "your_openai_api_key_here")

# Initialize the OpenAI LLM
llm = ChatOpenAI(
    model="gpt-4",  # or "gpt-3.5-turbo" for cost savings
    temperature=0.7,
    api_key=openai_api_key
)

print(f"âœ… OpenAI LLM initialized with model: gpt-4")
print(f"âš ï¸  Make sure to set your OPENAI_API_KEY in the .env file!")
```

## 3. Define the Engineering Lead Agent

The Engineering Lead is responsible for project planning, architecture decisions, and coordinating the team.

```python
engineering_lead = Agent(
    role='Engineering Lead',
    goal='Plan and coordinate the engineering team to deliver high-quality software solutions',
    backstory="""You are an experienced Engineering Lead with 15+ years in software development.
    You excel at breaking down complex projects into manageable tasks, making architectural decisions,
    and ensuring the team works cohesively. You have deep knowledge of software design patterns,
    system architecture, and project management. You prioritize clean code, scalability, and maintainability.""",
    verbose=True,
    allow_delegation=True,
    llm=llm
)

print("âœ… Engineering Lead agent created successfully!")
```

## 4. Define the Backend Engineer Agent

The Backend Engineer specializes in server-side development, database design, and API implementation.

```python
backend_engineer = Agent(
    role='Backend Engineer',
    goal='Design and implement robust, scalable backend systems and APIs',
    backstory="""You are a skilled Backend Engineer with expertise in RESTful API development,
    database design, microservices architecture, and server-side programming. You have experience
    with Python (FastAPI, Django, Flask), Node.js, and various databases (PostgreSQL, MongoDB, Redis).
    You focus on performance optimization, security best practices, and creating maintainable code.
    You understand cloud services, containerization, and DevOps principles.""",
    verbose=True,
    allow_delegation=False,
    llm=llm
)

print("âœ… Backend Engineer agent created successfully!")
```

## 5. Define the Frontend Engineer Agent

The Frontend Engineer creates user interfaces and implements client-side functionality.

```python
frontend_engineer = Agent(
    role='Frontend Engineer',
    goal='Create intuitive, responsive, and high-performance user interfaces',
    backstory="""You are an expert Frontend Engineer with deep knowledge of modern web technologies.
    You specialize in React, Vue.js, and Angular, along with HTML5, CSS3, and JavaScript/TypeScript.
    You have a strong eye for design and UX principles, ensuring applications are not only functional
    but also beautiful and user-friendly. You're experienced with state management (Redux, Vuex),
    responsive design, accessibility standards (WCAG), and frontend performance optimization.
    You keep up with the latest frontend trends and best practices.""",
    verbose=True,
    allow_delegation=False,
    llm=llm
)

print("âœ… Frontend Engineer agent created successfully!")
```

## 6. Define the Test Engineer Agent

The Test Engineer focuses on quality assurance, test strategy, and automated testing.

```python
test_engineer = Agent(
    role='Test Engineer',
    goal='Ensure software quality through comprehensive testing strategies and automation',
    backstory="""You are a meticulous Test Engineer with expertise in QA methodologies and test automation.
    You have experience with various testing frameworks (Jest, Pytest, Selenium, Playwright, Cypress) and understand
    different testing types: unit tests, integration tests, end-to-end tests, and performance tests.
    You're skilled at identifying edge cases, creating comprehensive test plans, and implementing CI/CD
    testing pipelines. You believe in shift-left testing and work closely with developers to catch
    bugs early. You're passionate about code coverage, test-driven development (TDD), and maintaining
    high software quality standards.""",
    verbose=True,
    allow_delegation=False,
    llm=llm
)

print("âœ… Test Engineer agent created successfully!")
```

## 7. Define Tasks for Each Agent

Now we'll create specific tasks for each agent to work on a sample project: building a task management application.

```python
# Task 1: Engineering Lead - Project Planning and Architecture
task_plan_architecture = Task(
    description="""Create a comprehensive project plan and architecture design for a Task Management Application.
    
    Requirements:
    - Define the overall system architecture (frontend, backend, database)
    - Identify key features: user authentication, task CRUD operations, task assignments, deadlines, priorities
    - Choose appropriate technology stack
    - Define API endpoints structure
    - Create a high-level project timeline
    - Identify potential risks and mitigation strategies
    
    Deliverable: A detailed architecture document with system design, tech stack recommendations, and project plan.""",
    agent=engineering_lead,
    expected_output="A comprehensive architecture document including system design, technology choices, API structure, and project timeline."
)

# Task 2: Backend Engineer - API Design and Database Schema
task_backend_design = Task(
    description="""Design the backend system for the Task Management Application based on the architecture plan.
    
    Requirements:
    - Design a RESTful API with all necessary endpoints (auth, tasks, users)
    - Create a detailed database schema (users, tasks, assignments, etc.)
    - Define data models and relationships
    - Plan authentication and authorization strategy (JWT, OAuth)
    - Consider scalability and performance optimizations
    - Define error handling and validation strategies
    
    Deliverable: Complete API specification with endpoint documentation and database schema design.""",
    agent=backend_engineer,
    expected_output="Detailed API specification with endpoints, request/response formats, and complete database schema with relationships."
)

# Task 3: Frontend Engineer - UI/UX Design and Component Structure
task_frontend_design = Task(
    description="""Design the frontend architecture and user interface for the Task Management Application.
    
    Requirements:
    - Create wireframes/mockups for key screens (dashboard, task list, task details, login)
    - Define component hierarchy and state management approach
    - Plan responsive design strategy (mobile, tablet, desktop)
    - Design user workflows (creating tasks, assigning tasks, filtering, searching)
    - Choose UI framework and component library
    - Plan accessibility features (ARIA labels, keyboard navigation)
    
    Deliverable: Frontend architecture document with component structure, UI mockups, and state management plan.""",
    agent=frontend_engineer,
    expected_output="Complete frontend architecture with component hierarchy, UI design specifications, and state management strategy."
)

# Task 4: Test Engineer - Testing Strategy and Test Plan
task_testing_strategy = Task(
    description="""Develop a comprehensive testing strategy for the Task Management Application.
    
    Requirements:
    - Create a test plan covering unit, integration, and e2e tests
    - Define test cases for critical user flows (login, create task, assign task, update task)
    - Plan automated testing strategy and CI/CD integration
    - Identify edge cases and error scenarios to test
    - Define performance testing approach
    - Plan security testing (authentication, authorization, input validation)
    - Set code coverage targets
    
    Deliverable: Comprehensive test plan with test cases, automation strategy, and quality assurance guidelines.""",
    agent=test_engineer,
    expected_output="Detailed test plan including test cases, automation strategy, and quality metrics for all application components."
)

print("âœ… All tasks defined successfully!")
```

## 8. Configure the Crew

Now we'll assemble all agents and tasks into a crew that will work together sequentially.

```python
# Create the crew with all agents and tasks
engineering_crew = Crew(
    agents=[
        engineering_lead,
        backend_engineer,
        frontend_engineer,
        test_engineer
    ],
    tasks=[
        task_plan_architecture,
        task_backend_design,
        task_frontend_design,
        task_testing_strategy
    ],
    process=Process.sequential,  # Tasks will be executed in order
    verbose=True
)

print("âœ… Engineering Crew assembled successfully!")
print(f"   - {len(engineering_crew.agents)} agents")
print(f"   - {len(engineering_crew.tasks)} tasks")
print(f"   - Process: Sequential execution")
```

## 9. Execute the Crew Workflow

Let's kick off the crew and watch our AI engineering team collaborate!

```python
# Execute the crew workflow
# This will take several minutes as each agent works on their task
print("ğŸš€ Starting the Engineering Crew workflow...")
print("â° This may take a few minutes as each agent completes their task.\n")

result = engineering_crew.kickoff()

print("\nâœ… Crew execution completed!")
```
